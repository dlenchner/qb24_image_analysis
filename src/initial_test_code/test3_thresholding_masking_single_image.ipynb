```python
import imageio.v2 as imageio
import matplotlib.pyplot as plt
import numpy as np
import skimage as ski
```

```python
image1 = imageio.imread(uri = "~/qb24_image_analysis/data/FluorescentCells.jpg")

fig, ax = plt.subplots()
ax.imshow(image1)
```

```python
image1_gray = ski.color.rgb2gray(image1)
image1_blur = ski.filters.gaussian(image1_gray, sigma = 1.0)

fig, ax = plt.subplots()
ax.imshow(image1_blur, cmap = "gray")
```

```python
histogram, bin_edges = np.histogram(image1_blur, bins=256, range=(0.0, 1.0))

fig, ax = plt.subplots()
ax.plot(bin_edges[0:-1], histogram)
ax.set_title("Grayscale Histogram")
ax.set_xlabel("grayscale value")
ax.set_ylabel("pixels")
ax.set_xlim(0, 0.5)
```

```python
t = 0.2
binary_mask = image1_blur > t

fig, ax = plt.subplots()
ax.imshow(binary_mask, cmap="gray")
```

```python
selection = image1.copy()
selection[~binary_mask] = 0

fig, ax = plt.subplots()
ax.imshow(selection)
```

```python
pixels = np.count_nonzero(binary_mask)
w = binary_mask.shape[1]
h = binary_mask.shape[0]

print(w*h)

print(pixels)

density = pixels / (w*h)

print(density)
```
